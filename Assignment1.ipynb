```python
#union
mySet = {1,5,7,8,5,6,9}
set1 = {20,34,7,5,9,13}
set2 = {34,80,66,30,52}
print(mySet.union(set1).union(set2))
```

```python
print(mySet | set1 | set2)
```

```python
#intersection
intersection_set = set1.intersection(set2)
print(intersection_set)
```

```python
#differnce
set3 = {1,2,6,8,10}
set4 = {2,4,6,8,10}
print(set3.difference(set4))
```
